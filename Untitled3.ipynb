## Section 5 — Medical Relationship Rules → Enriched Dataset

This section applies **plain arithmetic formulas** directly to the base dataset columns.  
Each rule encodes one documented physiological relationship.  
The correlations appear automatically because the formulas causally connect features.

### How each rule creates a correlation

For example, Rule 1 says:

```
systolic_bp = systolic_bp + 0.5 × (bmi - 25)
```

When BMI is high, SBP goes up by a proportional amount.  
When BMI is low, SBP goes up less (or not at all).  
This directly causes BMI and SBP to move together → correlation emerges.

Every rule below works the same way: a driver feature appears on the right-hand side  
of an equation that modifies a target feature.

### Rules applied

| # | Formula (plain English) | Medical Justification | Source |
|---|---|---|---|
| R1 | SBP += 0.5 × (BMI − 25) when BMI > 25 | Each BMI unit above 25 raises SBP ~0.5 mmHg via sympathetic activation and increased cardiac output | [AHA Hypertension — Obesity and Hypertension](https://www.ahajournals.org/doi/10.1161/HYPERTENSIONAHA.116.09004) |
| R2 | SBP += 0.4 × (age − 40) when age > 40 | Arterial stiffening with ageing raises systolic pressure ~0.4 mmHg per year after age 40 | [JACC — Arterial Stiffness and Hypertension](https://www.jacc.org/doi/10.1016/j.jacc.2019.12.009) |
| R3 | fasting_glucose += 15 × diabetes | Each diabetes severity level (pre-DM / T2 / T1) adds ~15 mg/dL fasting glucose — severity dose-response | [ADA Standards of Medical Care in Diabetes](https://diabetesjournals.org/care/article/46/Supplement_1/S1/148039) |
| R4 | LDL += 12 × smoking | Each smoking level adds ~12 mg/dL LDL — nicotine reduces HDL and raises oxidised LDL | [CDC — Smoking and Cardiovascular Disease](https://www.cdc.gov/tobacco/data_statistics/sgr/50th-anniversary/pdfs/fs_smoking_CVD_508.pdf) |
| R5 | BNP += 10 × (50 − EF) when EF < 50 | Each EF unit below 50% adds ~10 pg/mL BNP due to ventricular wall stretch releasing natriuretic peptide | [JACC — BNP and Left Ventricular Dysfunction](https://www.jacc.org/doi/10.1016/S0735-1097%2802%2901882-8) |
| R6 | BNP += 1.5 × (age − 40) when age > 40 | BNP rises ~1.5 pg/mL per year above 40 due to age-related cardiac fibrosis and remodelling | [Circulation Heart Failure — Age and BNP](https://www.ahajournals.org/doi/10.1161/CIRCHEARTFAILURE.119.006541) |
| R7 | BNP += 200 when SOB == 3 AND edema == 2 | Decompensated Heart Failure (severe dyspnoea + severe oedema) causes acute BNP spike of 100–500 pg/mL | [JACC — Acute Decompensated Heart Failure](https://www.jacc.org/doi/10.1016/j.jacc.2017.11.006) |
| R8 | BNP += 8 × (HR − 100) when HR > 100 | Sustained tachycardia stretches the atrial wall, triggering BNP release; each bpm above 100 adds ~8 pg/mL | [European Heart Journal — Tachycardia and BNP](https://academic.oup.com/eurheartj/article/24/19/1732/401777) |
| R9 | SpO2 −= 0.3 × (35 − EF) when EF < 35 | Severe HFrEF (EF < 35%) causes pulmonary congestion and impaired alveolar gas exchange → hypoxaemia | [NEJM — Heart Failure and Pulmonary Congestion](https://www.nejm.org/doi/10.1056/NEJMra1404688) |
| R10 | HR += 10 × (SOB − 1) when SOB ≥ 2 | Moderate to severe respiratory distress activates the sympathetic nervous system, driving compensatory tachycardia | [MedlinePlus — Shortness of Breath and Heart Rate](https://medlineplus.gov/ency/article/003075.htm) |
| R11 | fasting_glucose += 3 × (BMI − 30) when BMI > 30 | Visceral adiposity above BMI 30 drives insulin resistance; each excess BMI unit raises fasting glucose ~3 mg/dL | [Diabetes Care — Obesity and Insulin Resistance](https://diabetesjournals.org/care/article/31/Supplement_2/S261/28519) |
| R12 | LDL += 0.8 × (age − 40) when age > 40 | Hepatic VLDL production increases with age; each year above 40 raises LDL ~0.8 mg/dL | [Atherosclerosis — Age and LDL Cholesterol](https://www.atherosclerosis-journal.com/article/S0021-9150(06)00444-1/fulltext) |
